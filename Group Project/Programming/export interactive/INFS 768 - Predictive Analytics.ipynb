```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
```

```python
X = np.array([1,3,5,7,6])
```

```python
# scale numeric data with a scale factor
from logging import error
from pandas import DataFrame, core
def scale(df:core.frame.DataFrame, scaleFactor:int) -> dict:
    if(type(df)!=core.frame.DataFrame):
        return error(msg="Error! variable passed to the 'dataframe' parameter is not a dataframe")
    min = df.min()
    max = df.max()
    df = scaleFactor * (df - min) / (max - min)
    return dict(df=df, min=min, max=max, scaleFactor=scaleFactor)
def scale_inverse(scaleItem:dict) -> core.frame.DataFrame:
    min = scaleItem.get("min")
    max = scaleItem.get("max")
    scaleFactor = scaleItem.get("scaleFactor")
    df = scaleItem.get("df")
    return df*(max-min)/scaleFactor + min
```

```python
print(X)
display(np.array([X,X+1]))
display(np.transpose(np.array([X,X+1])))
```

```python

from sklearn.utils import shuffle


df = pd.DataFrame(np.transpose([X,X+1]), columns=["row1","row2"])
display(df)
scaleItem = scale(df, 10)
scaleItem2 = scale(shuffle(df), 10)
randoDF = scale_inverse(scaleItem)
display(randoDF)
display(scale_inverse(scaleItem2))
```
